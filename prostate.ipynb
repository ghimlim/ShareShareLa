```r
library(leaps)
leapsStat <- regsubsets(lpsa~., p2a, nvmax=8)
leapsStat.summary <- summary(leapsStat)
leapsStat.summary$outmat
```

```r
leapsStat.adjr2 <- which.max(leapsStat.summary$adjr2)
```

```r
leapsStat.adjr2
```

```r
leapsStat.cp <- which.min(leapsStat.summary$cp)
```

```r
leapsStat.bic <- which.min(leapsStat.summary$bic)
```
